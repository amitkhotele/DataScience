
### Question 1: What is hypothesis testing in statistics?
 Answer:
 Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data. It involves formulating a null hypothesis and an alternative hypothesis, then using sample data to determine whether to reject the null hypothesis.

### Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
 Answer:
 The null hypothesis (H0) is a statement that there is no effect or no difference, and it serves as the default assumption. The alternative hypothesis (H1 or Ha) is what you want to prove, indicating the presence of an effect or difference. Hypothesis testing determines whether there is enough evidence to reject H0 in favor of Ha.

### Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
 Answer:
 The significance level (alpha, α) is the probability of rejecting the null hypothesis when it is actually true (Type I error). Common values are 0.05 or 0.01. If the p-value from the test is less than α, we reject the null hypothesis.

### Question 4: What are Type I and Type II errors? Give examples of each.
 Answer:
 Type I error: Rejecting the null hypothesis when it is true (false positive). Example: Concluding a drug works when it actually doesn't.
 Type II error: Failing to reject the null hypothesis when it is false (false negative). Example: Concluding a drug doesn't work when it actually does.

### Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.
 Answer:
 A Z-test is used when the population variance is known and/or the sample size is large (n > 30). A T-test is used when the population variance is unknown and the sample size is small (n ≤ 30). Both test means, but the T-test accounts for extra uncertainty in small samples.


```python
# Question 6: Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.
import matplotlib.pyplot as plt
import numpy as np

n = 10
p = 0.5
size = 1000
binom_data = np.random.binomial(n, p, size)

plt.hist(binom_data, bins=range(n+2), edgecolor='black', align='left')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.show()
```

```python
# Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python.
import scipy.stats as stats

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Assume population mean = 50, population std = 1 (for Z-test)
pop_mean = 50
pop_std = 1
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)
z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print(f"Sample mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation:
# If p-value < 0.05, reject the null hypothesis. Otherwise, fail to reject.
```

```python
# Question 8: Simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data.
np.random.seed(0)
data = np.random.normal(loc=100, scale=15, size=200)
mean = np.mean(data)
std_err = stats.sem(data)
conf_int = stats.norm.interval(0.95, loc=mean, scale=std_err)

plt.hist(data, bins=20, edgecolor='black')
plt.title('Simulated Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.axvline(conf_int[0], color='red', linestyle='dashed', label='95% CI Lower')
plt.axvline(conf_int[1], color='green', linestyle='dashed', label='95% CI Upper')
plt.legend()
plt.show()

print(f"Sample mean: {mean:.2f}")
print(f"95% Confidence Interval: {conf_int}")
```

```python
# Question 9: Python function to calculate Z-scores and visualize standardized data using a histogram.
def plot_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    plt.hist(z_scores, bins=20, edgecolor='black')
    plt.title('Histogram of Z-scores')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.show()
    return z_scores

# Example usage:
z_scores = plot_z_scores(np.array(sample_data))

# Explanation:
# Z-scores represent how many standard deviations a data point is from the mean. A Z-score of 0 means the value is at the mean, positive values are above the mean, and negative values are below.
```
