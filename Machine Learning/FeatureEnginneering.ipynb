# Assignment Questions

1. **What is a parameter?**
    - A parameter is a variable in a model that is learned from data during training, such as weights in linear regression.

2. **What is correlation?**
    - Correlation measures the relationship between two variables, indicating how one variable changes with respect to another.

3. **What does negative correlation mean?**
    - Negative correlation means that as one variable increases, the other decreases.

4. **Define Machine Learning. What are the main components in Machine Learning?**
    - Machine Learning is a field of artificial intelligence that enables systems to learn from data and improve over time without explicit programming. Main components include Data, Model, Loss Function, and Optimizer.

5. **How does loss value help in determining whether the model is good or not?**
    - The loss value quantifies the difference between predicted and actual values. Lower loss values indicate better model performance; high loss means poor predictions.

6. **What are continuous and categorical variables?**
    - Continuous variables can take any numerical value within a range (e.g., height, weight). Categorical variables represent discrete categories or groups (e.g., gender, color).

7. **How do we handle categorical variables in Machine Learning? What are the common techniques?**
    - Categorical variables are converted to numerical format using techniques like Label Encoding, One-Hot Encoding, and Ordinal Encoding.

8. **What do you mean by training and testing a dataset?**
    - The training dataset is used to fit the model, while the testing dataset is used to evaluate the model's performance on unseen data.

9. **What is sklearn.preprocessing?**
    - `sklearn.preprocessing` is a module in scikit-learn for data preprocessing, including scaling, encoding, and transformation.

10. **What is a Test set?**
    - A test set is a subset of data used to assess the performance of a trained model and estimate its generalization ability.

11. **How do we split data for model fitting (training and testing) in Python?**
    - Data can be split using `train_test_split` from `sklearn.model_selection`, which divides data into training and testing sets.

12. **How do you approach a Machine Learning problem?**
    - Steps: Understand the problem, collect data, preprocess data, perform exploratory data analysis (EDA), engineer features, select a model, train the model, evaluate performance, and deploy the solution.

13. **Why do we have to perform EDA before fitting a model to the data?**
    - EDA helps understand data distribution, detect anomalies, identify patterns, and guide preprocessing and feature engineering.

14. **What is correlation?**

15. **What does negative correlation mean?**
    - Negative correlation means that as one variable increases, the other decreases.

16. **How can you find correlation between variables in Python?**
    - You can use the `df.corr()` method in pandas to compute correlation between variables in a DataFrame.

17. **What is causation? Explain difference between correlation and causation with an example.**
    - Causation means one variable directly affects another. Correlation is just association. Example: Ice cream sales and drowning rates are correlated (both increase in summer), but eating ice cream does not cause drowning.

18. **What is an Optimizer? What are different types of optimizers? Explain each with an example.**
    - An optimizer updates model parameters to minimize the loss function. Types:
        - **SGD (Stochastic Gradient Descent):** Updates weights using the gradient of the loss for each sample.
        - **Adam:** Uses adaptive learning rates and combines momentum and RMSprop techniques.
        - **RMSprop:** Maintains per-parameter learning rates, adjusting them based on recent gradients.

19. **What is sklearn.linear_model ?**
    - `sklearn.linear_model` is a module in scikit-learn for linear models such as LinearRegression and LogisticRegression.

20. **What does model.fit() do? What arguments must be given?**
    - `model.fit()` trains the model on the given data. Arguments: features (X) and target (y).

21. **What does model.predict() do? What arguments must be given?**
    - `model.predict()` predicts target values for new data. Argument: features (X).

25. **What is sklearn.preprocessing?**
